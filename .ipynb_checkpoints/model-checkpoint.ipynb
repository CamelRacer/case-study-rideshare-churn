```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_curve
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/churn_train.csv')
df['churn'] = df['last_trip_date'] < '2014-06-01'
bool_map = {True: 1, False: 0}
df.churn = df.churn.map(bool_map)
df.luxury_car_user = df.luxury_car_user.map(bool_map)
df.head()
```

```python
null_avgdist = df[(df.avg_dist == 0) & (df.trips_in_first_30_days > 0)].index
null_avgrateby = df[df.avg_rating_by_driver.isnull()].index
null_avgrateof = df[df.avg_rating_of_driver.isnull()].index
null_phone = df[df.phone.isnull()].index

df2 = df.drop(null_avgdist)
df2 = df.drop(null_avgrateby)
df2 = df.drop(null_phone)

df2.avg_rating_of_driver = df2.avg_rating_of_driver.fillna(avgrateof_mean)
drop = df.drop(null_avgrateof)
avgrateof_mean = np.mean(drop.avg_rating_of_driver)

df2.count()
```

```python
df2.signup_date = pd.to_datetime(df2.signup_date)
df2.last_trip_date = pd.to_datetime(df2.last_trip_date)
df2.signup_age = pd.to_datetime('2014-07-01') - df2.signup_date
df2.last_trip_age = pd.to_datetime('2014-07-01') - df2.last_trip_date

df2 = pd.get_dummies(df2)
df2
```

```python
y = df2.churn
X = df2.drop(['churn', 'last_trip_date', 'signup_date', 'avg_rating_by_driver'], axis=1)
#subset selection (same accuracy as before)
# X = X.drop(['phone_Android', 'phone_iPhone', 'luxury_car_user', 'avg_rating_of_driver'], axis=1)
#2nd subset selection
# X = X.drop(['city_Astapor', 'city_Winterfell', 'avg_surge'], axis=1)

X_train, X_val, y_train, y_val = train_test_split(X, y)
```

```python
df3 = df2[df2['city_Astapor'] == 1]
df3.head()

y = df3.churn
X = df3.drop(['churn', 'last_trip_date', 'signup_date', 'avg_rating_by_driver'], axis=1)
#subset selection (same accuracy as before)
# X = X.drop(['phone_Android', 'phone_iPhone', 'luxury_car_user', 'avg_rating_of_driver'], axis=1)
#2nd subset selection
X = X.drop(['city_King\'s Landing', 'city_Winterfell'], axis=1)

X_train, X_val, y_train, y_val = train_test_split(X, y)
```

```python
param_grid = {'n_estimators': [10, 100],
              'max_features': ['sqrt', 'log2', None],
              'max_depth': [2, 5, 10, 20],
              'random_state': [1]}

model_rf = RandomForestClassifier()
grid = GridSearchCV(model_rf, param_grid, n_jobs=-1, verbose=2)
grid.fit(X_train, y_train)
```

```python
# grid.best_params_
# zip(grid.best_estimator_.feature_importances_, X_train.columns.values)
grid.best_score_
```

```python
ada_param = {'learning_rate': [.5, .75, .9, 1],
             'random_state': [1],
             'n_estimators': [200, 300, 500]}

ada = AdaBoostClassifier()
models_ada = GridSearchCV(ada, ada_param, n_jobs=-1, verbose=1)
models_ada.fit(X_train, y_train)

```

```python
models_ada.best_score_
```

```python
zip(models_ada.best_estimator_.feature_importances_, X_train.columns.values)
```

```python
recall_score(y_val, models_ada.best_estimator_.predict(X_val))
```

```python
precision_score(y_val, models_ada.best_estimator_.predict(X_val))
```
